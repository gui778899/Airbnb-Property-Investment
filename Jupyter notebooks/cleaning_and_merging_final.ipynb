The function below clean_bathroom_value converts the number of bathrooms to float-type and the catch the attibute error for those value that are null. These null values are left as they null to later deal with them with another function.

```python
def clean_bathroom_value(x):
  try:
    index = x.index(" ")
    return float(x[:index])
  except ValueError:
    return 1
  except AttributeError:   #those are null values, it will just keept the null values for now
    return np.nan
```

The function below checks if there are words in the bathrooms values. If yes it returns one. This function is  will be used as a lambda function in later stages to count the Private, Shared and Half bathroom.

```python
def check_words_in_bathroom_values(x, words):
  try:
    for word in words:
      if word in x: return 1
    return 0
  except TypeError: return np.nan
```

It is imported from the geopy python library (Python client for several popular geocoding web services) to find the geographical center point of a given uk postcode sector by converting it into latitude and longitude cordinates using teh OpenStreetMap's Nominatim service.

```python
from geopy.geocoders import Nominatim

def get_centroid(postcode_sector):
    geolocator = Nominatim(user_agent="postcode_centroid")

    # Assume the postcode_sector is a valid UK postcode sector (e.g., "SW1A 1")
    location = geolocator.geocode(postcode_sector)

    if location:
        return location.latitude, location.longitude
    else:
        return None
```

## Matching postcode with neighbourhoods

**Data**: There are two datasets:

**Airbnb Dataset**: Contains neighborhoods with latitude and longitude but lacks postcode information.

**Postcodes Provisional Dataset**: Contains postcodes and their corresponding centroid coordinates (latitude and longitude).

**Objective**: Want to find out which postcode sector each neighborhood is closest to.

**Method**: For each neighborhood, calculate the Euclidean distance to each postcode sector's centroid using the euclidian_dist_sq function. Then, compare these distances to find the smallest one, which indicates the closest postcode sector.

**Assignment**: Once it is found the smallest distance for a neighborhood, it is  assigned the corresponding postcode sector to that neighborhood.

**Result**: The outcome is a mapping from neighborhoods to postcode sectors, based on proximity to the centroid coordinates of those sectors.

```python
def euclidian_dist_sq(tup1, tup2):
  return (tup1[0]-tup2[0])**2 + (tup1[1]-tup2[1])**2  #


def get_closest_area_code(row):
  coords = row["latitude"], row["longitude"]
  min_dist = 100000000
  best_match = "None"
  for code in coords_dico:
    dist = euclidian_dist_sq(coords, coords_dico[code])
    if dist<min_dist:
      best_match = code
      min_dist = dist

  return best_match
```

Note: In this analysis,  Euclidean distance is utilised.
The square root operation required for calculating the actual Euclidean distance
adds computational complexity. When the objective is solely to compare relative distances
to determine the closest point, the square root component is unnecessary.
Squared distances maintain the comparative relationships between points,
making them a suitable substitute for distance comparison. This optimization
omits the computationally intensive square root calculation, thereby enhancing performance
while preserving the accuracy of the comparative analysis.

As the linear regression is not suitable to predict the price of the NaN rooms, is was decided to adopt a more reliable method adopting a linear interpolation as a more robust solution as seen by the code below.

```python
def estimate_price(row):
  postcode = row["postcode_estimate"]
  rooms = row["bedrooms"]+ row["bathrooms_count"]

  postcode_only = late[late["Outward code"] == postcode]
  exact_room_count = postcode_only[postcode_only["Number of rooms"] == rooms]

  if len(exact_room_count)>0:
    return exact_room_count["Price"].iloc[0]

  #Now if the exact room count is absent from the dataframe. We Will perform linear interpolation between the two closest room counts
  room_count_above = postcode_only[postcode_only["Number of rooms"] > rooms]["Number of rooms"].min()
  room_count_below = postcode_only[postcode_only["Number of rooms"] < rooms]["Number of rooms"].max()

  if math.isnan(room_count_above):   #This is the case when there is no data for houses with more rooms than the target
    price_below=postcode_only[postcode_only["Number of rooms"] == room_count_below]["Price"].iloc[0]
    return rooms* price_below / room_count_below

  price_above=postcode_only[postcode_only["Number of rooms"] == room_count_above]["Price"].iloc[0]
  price_below=postcode_only[postcode_only["Number of rooms"] == room_count_below]["Price"].iloc[0]

  return price_below+(rooms-room_count_below)/(room_count_above - room_count_below)*(price_above - price_below)
```

## Estimate the time it takes a house to recoup its original value

If the house price and the price per night for resting were explicit would be easier to calculate how much a house makes per night/week/month etc.. So the objective of this section is to estimate a house delay until it returns its original value.

To do so, it is needed to estimate how much each property is rented, this a crucial aspect of this research project since the ratio price/renting_price does not mean anything if the property is not actually being rented.

The best way to perform the estimation how many days the property is actually being rented is using the column availability_90, as it can be computed the number of days a property was rented based on how much it was available for rent and therefore not already rented. (90 days - rental days) = not rented days out of 90 days

```python
df["availability_365"].hist()
```

Create the rented_days column that compute the rented days the property is actually rented

```python
df["rented_days"] = 90 - df["availability_90"]
```

If it is multiplied the renting price by the rented_days, it is possible to approximate how much a house makes in three months (90 days).

This allows to determine, in months, how much does a house take to yield its original value. The rented_days were first set to a minimium of 1, since if was set to 0 there would be a erro in this function because it is impossible to divide by 0.

```python
df.loc[df['rented_days'] == 0, 'rented_days'] = 1
```

To calculate the delay it is applied the formula below:

```python
df["delay"] = df["price_estimate"]/(df["price"]* df["rented_days"])
```
